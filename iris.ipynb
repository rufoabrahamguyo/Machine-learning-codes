```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
iris=load_iris()
iris_df=pd.DataFrame(np.c_[iris.data,iris.target],columns=np.append(iris.feature_names,['target']))
iris.keys()
```

```python
grid=pd.plotting.scatter_matrix(iris_df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X_train,X_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.2,random_state=42)
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
print("Predicted values:",y_pred)
print("Actual values: ",y_pred)
print("Accuracy {:.2f}".format(np.mean(y_pred==y_test)))
print("Accuracy based on score function {:.2f}".format(knn.score(X_test,y_test)))
```

```python
prediction=knn.predict([[4,3,10,5]])
```

```python
print("make new predictions",prediction)
```

```python
print("predicted class name ",iris['target_names'][prediction])
```
